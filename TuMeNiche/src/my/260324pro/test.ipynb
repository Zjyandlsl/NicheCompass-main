```python
import urllib.request
import time
import ssl

# 忽略 SSL 证书警告（如果有）
ssl._create_default_https_context = ssl._create_unverified_context

# 这里列出 NicheNet 常用的几个下载源地址（测试用的小文件）
test_urls = [
    "https://zenodo.org/record/7074291/files/nichenet_lr_network.csv",  # 示例地址
    "https://raw.githubusercontent.com/saeyslab/nichenetr/master/data-raw/nichenet_lr_network.csv"
]

def test_download_speed(url):
    print(f"\n正在测试连接: {url[:60]}...")
    try:
        start_time = time.time()
        
        # 只下载前 100KB 来测试速度，不下载全部
        req = urllib.request.Request(url, headers={'Range': 'bytes=0-102400'})
        with urllib.request.urlopen(req, timeout=10) as response:
            data = response.read()
            elapsed = time.time() - start_time
            speed_kb_s = (len(data) / 1024) / elapsed
            
            print(f"   ✅ 连接成功！")
            print(f"   耗时: {elapsed:.2f} 秒")
            print(f"   速度: {speed_kb_s:.2f} KB/s")
            
            if speed_kb_s < 50:
                print("   ⚠️  结论：网速很慢（低于 50KB/s），这很可能是瓶颈。")
            else:
                print("   ✅ 结论：网速正常。")
                
    except Exception as e:
        print(f"   ❌ 连接失败: {e}")

# 执行测试
for url in test_urls:
    test_download_speed(url)
```
